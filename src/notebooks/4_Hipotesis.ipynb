## Importamos librer√≠as necesarias

```python
import pandas as pd

import scipy.stats as stats
import sys

pd.set_option("display.max_columns", None)
sys.path.append("../")
```

## Pre-cargamos datasets

```python
df_torneos = pd.read_csv("../data/final/resultados_final.csv")
df_torneos['Date'] = pd.to_datetime(df_torneos['Date'])
df_torneos.head()
```

```python
dfStats = pd.read_csv("../data/final/estadisticas_final.csv")
dfStats.drop("Unnamed: 0", axis=1, inplace=True)
dfStats.head()
```

## Hip√≥tesis

### Hip√≥tesis 1
- H0 -> No hay evidencia significativa de la correlaci√≥n entre n√∫mero de aces y el n√∫mero de servicios ganados
- H1 -> Existe una evidencia significativa de la correlaci√≥n entre n√∫mero de aces y el n√∫mero de servicios ganados

```python
df_aces = dfStats["Aces"]
df_DobleFalta = dfStats["Service Games Won"]

correlacion , pv = stats.pearsonr(df_aces, df_DobleFalta)

if pv < 0.05:
    print('Rechazamos la hip√≥tesis nula "No hay evidencia significativa de la correlaci√≥n entre n√∫mero de aces y el n√∫mero de servicios ganados"')
else:
    print('No rechazamos la hip√≥tesis nula')
```

### Hip√≥tesis 2
- H0 -> No hay evidencia significativa de la relaci√≥n entre n√∫mero de dobles faltas y el n√∫mero de Break Points salvados
- H1 -> Existe una evidencia significativa de la relaci√≥n entre n√∫mero de dobles faltas y el n√∫mero de Break Points salvados

```python
doblesFaltas = dfStats["Double Faults"]

breakPointSaved = dfStats["BreakPoints Saved"]

ttest = stats.ttest_ind(doblesFaltas, breakPointSaved)


if ttest[1] < 0.05:
    print('Rechazamos la hip√≥tesis nula "No hay evidencia significativa de la relaci√≥n entre n√∫mero de dobles faltas y el n√∫mero de Break Points salvados"')
else:
    print('No rechazamos la hip√≥tesis nula')
```

### Hip√≥tesis 3
- H0 -> La distribuci√≥n de las columnas num√©ricas sigue una distribuci√≥n normal
- H1 -> La distribuci√≥n de las columnas num√©ricas NO sigue una distribuci√≥n normal.

```python
#4.
variables_num = dfStats.select_dtypes(exclude='object')

shapiro = {var: stats.shapiro(variables_num[var]) for var in variables_num.columns}

for var in variables_num.columns:
    if shapiro[var][1] < 0.05:
        print(f'Rechazamos la hip√≥tesis nula "No hay evidencia significativa en la normalidad de la variable {var}"')
    else:
        print(f'No rechazamos la hip√≥tesis nula "No hay evidencia significativa en la normalidad de la variable {var}"')
```

### Hip√≥tesis 4
- H0 -> No hay evidencia significativa de la correlaci√≥n entre n√∫mero de TB jugados en hierba y el n√∫mero de TB jugados en arcilla
- H1 -> Existe una evidencia significativa de la correlaci√≥n entre n√∫mero de TB jugados en hierba y el n√∫mero de TB jugados en arcilla

```python
partidosHierba = df_torneos[df_torneos["Surface"] == "Grass"]["Tie_Breaks"]

partidosClay = df_torneos[df_torneos["Surface"] == "Clay"]["Tie_Breaks"]

ttest = stats.ttest_ind(partidosHierba, partidosClay)

if ttest[1] < 0.05:
    print('Rechazamos la hip√≥tesis nula "No hay evidencia significativa de la correlaci√≥n entre n√∫mero de TB jugados en hierba \ny el n√∫mero de TB jugados en arcilla"')
else:
    print('No rechazamos la hip√≥tesis nula')
```

### Hip√≥tesis 5
- H0 -> No hay evidencia de diferencias significativas en el n√∫mero de sets jugados en los Grand Slam, Masters 1000 y Copa de Maestros
- H1 -> Existe una evidencia de diferencias significativas en el n√∫mero de sets jugados en los Grand Slam, Masters 1000 y Copa de Maestros

```python
grupos_torneos = df_torneos.groupby('Series')['Sets_Played']
grandSlam = grupos_torneos.get_group('Grand Slam')
masters1000 = grupos_torneos.get_group('Masters 1000')
mastersCup = grupos_torneos.get_group('Masters Cup')

anova = stats.f_oneway(grandSlam, masters1000, mastersCup)
anova

if anova[1] < 0.05:
    print('Rechazamos la hip√≥tesis nula "No hay evidencia de diferencias significativas en el n√∫mero de sets jugados \nen los Grand Slam, Masters 1000 y Copa de Maestros"')
else:
    print('No rechazamos la hip√≥tesis nula')
```

### Hip√≥tesis 6
- H0 -> No hay evidencia significativa de la correlaci√≥n entre n√∫mero de sets jugados y el n√∫mero de Tie Breaks jugados
- H1 -> Existe una evidencia significativa de la correlaci√≥n entre n√∫mero de sets jugados y el n√∫mero de Tie Breaks jugados

```python
sets = df_torneos["Sets_Played"]
tieBreaks = df_torneos["Tie_Breaks"]

correlacion , pv = stats.pearsonr(sets, tieBreaks)

if pv < 0.05:
    print('Rechazamos la hip√≥tesis nula "No hay evidencia significativa de la correlaci√≥n entre n√∫mero de sets jugados y el n√∫mero de Tie Breaks jugados"')
else:
    print('No rechazamos la hip√≥tesis nula')
```
